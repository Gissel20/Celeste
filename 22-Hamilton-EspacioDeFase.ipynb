# Espacio de Fase

```python
from numpy import *
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import odeint
from numpy.linalg import norm
%matplotlib nbagg
```

# Hamiltoniano

```python
# Hamiltoniano del Problema 1-d equivalente de los dos cuerpos
def Unewton(r,params):
    mu=params["mu"]
    h=params["h"]
    U=0.5*r**2
    return U

def Hcentral(pr,r,params):
    U=params["U"]
    H=0.5*pr**2/(1+r**2)+U(r,params)
    return H
```

# Gráfico del Hamiltoniano en 3D

```python
params=dict(mu=1,h=1.5,U=Unewton)

fig = plt.figure()
ax=Axes3D(fig)

r=linspace(0.1,1,100)
pr=linspace(-10,10,100)

PR,R=meshgrid(pr,r)
H=Hcentral(PR,R,params)

ax.plot_surface(R,PR,H,rstride=2,cstride=2,cmap=cm.jet,lw=0,antialiased=True)
```

# Gráfico del Hamiltoniano en 2D

```python
fig=plt.figure()
ax=fig.gca()

params=dict(mu=1,h=0.4,U=Unewton)
r=linspace(-3,3,100)
pr=linspace(-3,3,100)
PR,R=meshgrid(pr,r)
H=Hcentral(PR,R,params)

c=ax.contour(R,PR,H,levels=linspace(-3,2,15))
ax.clabel(c)
```
